```python
#File Handling Assignment
print('1:Sign Up \n 2: Sign In \n 3: Forgot Password')
n=int(input('Enter your Input:'))

if n==1:#Sign Up

  import re

  username=input('Enter your Username: ')
  pattern=re.compile('[a-zA-Z0-9.,]+@+[a-zA-Z]+\.+[a-zA-Z]+')
  a=pattern.findall(username)

  flag=True
  if (a):
    print('Username is Valid')
    flag=True
  else:
    print('Username is not Valid')
    flag=False
  print(a)

  password=input('Enter your Password: ')
  for i in password:
    if len(password)>5 and len(password)<16:
      continue
    else:
      print('Length of password should be between 5 and 16.\n Enter Password Again.')
      break
  pattern1=re.compile('[a-zA-Z0-9]+[!@#$%^&*()]+[a-zA-Z0-9]+')
  b=pattern1.findall(password)
  
  flag=True
  if (b):
    print('Password is Valid')
    flag=True
  else:
    print('Password is not Valid')
    flag=False
  print(b)

  if flag==True:
    F1=open('Username1','a')
    F1.write('\n')
    F1.writelines(a)
    F1.close()
    F2=open('Password1','a')
    F2.write('\n')
    F2.writelines(b)
    F2.close()

if n==2: #Sign In
  import re

  flag=True
  username=input('Enter your Username: ')
  F1=open('Username1','r')
  for i in F1:
    if i==username:
      print('Valid User')
      flag=True
      break
  else:
    print('Invalid User')
    flag=False
      
  if flag==True:
    password=input('Enter your Password: ')
    F2=open('/content/Password1','r')
    for i in F2:
      if i==password:
        print('Logged In')
        break
    else:
        print('Invalid Password!! \n Try again ')
    F2.close()
  F1.close()

elif n==3: #forgot password
  
  import re
  
  username=input('Enter your Username: ')
  password=input('Enter your previous password: ')
  flag=True
  F1=open('/content/Username1','r')
  F2=open('/content/Password1','r+')
  for j in F1:
    if j==username:
      flag=True
      break
  else:
    flag=False

  if flag==True:
    flag=True
    for j in F2:
      if j==password:
        print('Password is valid')
        flag=True
        break
    else:
      print('Password is not valid \n Enter new password: ')
      flag=False

  if flag==False:
    password=input('Enter new password')
    result1=re.compile('[a-zA-Z0-9]+[!@#$%^&*()]+[a-zA-Z0-9]+')
    answer1=result1.findall(password)

  if (answer1):
    print('Password is Valid')
    F2.write('\n')
    F2.close()
  else:
    print('Password is not Valid')
  print(answer1)
  F1.close()
```
